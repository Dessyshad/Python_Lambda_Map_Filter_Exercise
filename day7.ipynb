## LAMBDA,MAP,FILTER FUNCTION

```python
# Lambda: It is an anonymous function that is designed to be used once.
# syntax: lambda args: operation to perform
area_rect = lambda l, b:l * b
area_rect(34,56)
```

```python
area_sq=lambda a:a**2
area_sq(10)
```

```python
even_odd=lambda x :f"{x} is an even number" if x % 2 == 0 else f"{x} ia an odd number"
even_odd(23)
```

## Map Function

```python
#
Numbers = [23,45,67,89,90]
even = lambda y : y % 2 == 0
print(list(map(even,Numbers)))
```

```python
print(list(map(even_odd, Numbers)))
```

```python
Animals = ["Lion","Tiger","Duck", "Eagle", "Elephant"]
display = lambda q : print(q)
print(list(map(display, Animals)))
```

```python
import statistics as stat
mean_value = lambda *num : stat.mean(num)
num1 = input("Enter the list of number to get the mean value:").split(",")
num = [float(i.strip()) for i in num1]
print(f"The mean of {num1}) is {mean_value(*num)}")
```

```python
#create a Tuple of Integers
#use the lambda function to perform a multiplication of 2
# use the map function to perform this lambda function on the Tuple of integers you created
info = (20,47,19,38)
operation = lambda m :m * 2 
print(tuple(map(operation, info)))
```

## Filter Function

```python
Numbers = [23,45,67,89,90]
even = lambda y : y % 2 == 0
print(list(filter(even, Numbers)))
```

```python
states = ["Oyo", "Ondo", "Yobe", "Imo", "Anambra"]
selection = lambda a: a.startswith("O")
print(list(filter(selection,states)))
```

```python
Word = "PyThOn"
#use the Lambda function to select all the letters that are capital letters
# use the filter function to display them
cap = lambda z: z.isupper()
print(list(filter(cap, Word)))
```

```python
cap = lambda z:z.islower()
print(list(filter(cap, Word)))
```
